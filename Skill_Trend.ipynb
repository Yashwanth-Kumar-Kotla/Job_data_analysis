```python
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading Data
df = pd.read_csv("https://huggingface.co/datasets/lukebarousse/data_jobs/resolve/main/data_jobs.csv")

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US_DA = df[(df["job_title_short"] == "Data Analyst") & (df["job_country"] ==  "United States")].copy()
```

```python
df_US_DA["job_posted_month"] = df_US_DA["job_posted_date"].dt.month
```

```python
df_US_DA_explode = df_US_DA.explode("job_skills")
```

```python
df_pivot = df_US_DA_explode.pivot_table(index="job_posted_month",columns="job_skills",aggfunc="size",fill_value=0)
```

```python
df_pivot.loc["total"] = df_pivot.sum()

df_pivot = df_pivot[df_pivot.loc["total"].sort_values(ascending=False).index]

df_pivot
```

```python
df_pivot.drop("total", inplace=True)
```

```python
DA_total = df_US_DA.groupby("job_posted_month").size()
```

```python
df_percent = df_pivot.div(DA_total, axis=0) * 100
```

```python
df_percent
```

```python
df_percent = df_percent.reset_index()
df_percent["job_posted_month_name"] = df_percent["job_posted_month"].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_percent = df_percent.set_index("job_posted_month_name")
df_percent = df_percent.drop(columns=["job_posted_month"])

df_percent
```

```python
df_percent.iloc[:,:5].plot(kind="line")
```

```python
df_plot = df_percent.iloc[:,:5]

sns.lineplot(data=df_plot,dashes= False, palette="tab10")
sns.set_theme(style="ticks")
sns.despine()
plt.title("Top 5 Skills Trend for Data Analyst in US")
plt.legend().remove()
plt.xlabel("2024")
plt.ylabel("Percentage of Job Postings")

for i in range(5):
    ax = plt.gca()
    ax.text(11.3, df_plot.iloc[-1, i], df_plot.columns[i])
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_plot = df_percent.iloc[:, :5]

sns.set_theme(style="ticks")
sns.lineplot(data=df_plot, dashes=False, palette="tab10")
sns.despine()

plt.title("Top 5 Skills Trend for Data Analyst in US")
plt.xlabel("2024")
plt.ylabel("Percentage of Job Postings")
plt.legend().remove()

# Get last values from each column
last_y_values = df_plot.iloc[-1].values
label_names = df_plot.columns

offsets = [0.5, 1.2, -1.2, 0.8, -0.8]

ax = plt.gca()
for i in range(5):
    y = last_y_values[i] + offsets[i]
    ax.text(11.3, y, label_names[i], fontsize=9, va='center')

plt.tight_layout()
```
